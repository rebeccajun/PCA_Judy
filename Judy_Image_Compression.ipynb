```python
from PIL import Image
im1 = Image.open('judy.jpeg')
im1.save('judy.png')
```

```python
import matplotlib.image as mplib
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
```

```python
img = mplib.imread('judy.png')
print(img.shape)
```

```python
plt.figure(figsize = (10,7))
plt.imshow(img);
```

```python
img_r = np.reshape(img,(2788,1534*3))
print(img_r.shape)
```

## n_components = 1 (~50% of variance)

```python
pca1 = PCA(n_components = 1).fit(img_r)
img_trans = pca1.transform(img_r)
print(np.sum(pca1.explained_variance_ratio_))
```

```python
# Reverse transform the reduced image to the 
# original size

rev1 = pca1.inverse_transform(img_trans)

# Let's see the image
# Firt we reshape it back into a 3D array
rev1 = np.reshape(rev1, (2788, 1534, 3))
f = plt.figure(figsize=(12,6)) # Create the canvas
ax1 = f.add_subplot(121)   # define plot1 on canvas
ax2 = f.add_subplot(122)   # define plot2 on canvas

ax1.imshow(img)
ax2.imshow(rev1)
ax1.set_title('Original Judy')
ax2.set_title('Transformed Judy, n_components=1')

plt.show()
```
